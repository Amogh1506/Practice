```python
import numpy as np

# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, IBMQ,assemble
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *
from qiskit.providers.aer import *

# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

```python
qc = QuantumCircuit(2)
#qc.save_statevector()
qc.x(0)
qc.x(1)
qc.h(1)

#qobj = assemble(qc)
qc.draw()
```

```python
sim = Aer.get_backend("statevector_simulator")
result = sim.run(qc).result()
state = result.get_statevector(qc)
counts = result.get_counts()
print(state)
print(counts)
plot_state_city(state)
# How to validate density matrix
# identify the end state of the circuit. in this case it will be (|1> * (1/sqrt(2)*(|0>-|1>)))
# do the full multiplication to get the complete state of the circuit |psi>: (1/sqrt(2)*(|10>-|11>))
# now...for pure states density matric is outter multiplication of |psi> * <psi|
# it becomes = 1/2 [(|10>-|11>) *(<10|-<11|) 
# 1/2[(|10>*<10|)-(|10><11|)-(|11>*<10|)+(|11>*<11|)]
# now chage the ket to columns and bra to rows e.g. <10| changes to [0010] and <11| changes to [0001]
# in the end check the complete matrix and convert to little endian qiskit ordering. u will see the exact above density matrix laid down





```
