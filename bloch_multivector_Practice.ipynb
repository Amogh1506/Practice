```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector,plot_histogram
from numpy import pi as pi
import matplotlib
```

```python
backend2= Aer.get_backend('statevector_simulator')
#qr = QuantumRegister(1)
#cr = ClassicalRegister(1)
qc = QuantumCircuit(1,1)
qc.ry(3*pi/4,0)
#qc.h(0)
res = execute(qc,backend = backend2).result()
#from qiskit.tools.visualization import plot_histogram
#plot_histogram(res.get_counts(qc))
statevector = res.get_statevector()
plot_bloch_multivector(statevector)
```

```python
qc.measure_all()
backend1 = Aer.get_backend('qasm_simulator')
res2 = execute(qc,backend = backend1).result()
plot_histogram(res2.get_counts(qc))
```
