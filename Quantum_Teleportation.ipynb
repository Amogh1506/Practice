```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3,3)
circuit.draw()
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw()
```

```python
circuit.h(1)
circuit.cx(1,2)
circuit.draw()
```

```python
circuit.cx(0,1)
circuit.h(0)
circuit.barrier()
circuit.draw()
```

```python
circuit.measure([0,1],[0,1])
circuit.barrier()
circuit.draw()
```

```python

circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw()
```

```python
circuit.measure(2,2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend= simulator,shots =1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
